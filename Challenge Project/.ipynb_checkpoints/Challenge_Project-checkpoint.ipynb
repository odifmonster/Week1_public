```python
# ---------------------------------------
# |     LOADING AND FORMATTING DATA     |
# ---------------------------------------
import pandas as pd
import numpy as np
import seaborn as sns

# load data from csv
url = "https://raw.githubusercontent.com/BeaverWorksMedlytics/datasets/master/allhypo.train.data.csv"
data_frame = pd.read_csv(url)

# replace '?' with NaN and drop empty columns if existing
data_frame[data_frame=='?'] = np.nan
try:
    data_frame = data_frame.drop(['TBG', 'TBG measured'],axis=1)
except:
    print("Trying to drop non-existing columns")

# convert numeric columns to floats
numeric_columns = ['Age', 'TSH', 'T3', 'TT4', 'T4u', 'FTI']
for label in numeric_columns:
    data_frame[label] = data_frame[label].apply(pd.to_numeric)

# replace NaN with median
data_frame = data_frame.fillna(data_frame.median())

# remove patient number from class column
data_frame['class'] = [data_frame['class'][x].split('.')[0] for x in data_frame.index]

# convert class to binary numeric values
data_frame['class'][data_frame['class']=='negative'] = 0.0
data_frame['class'][data_frame['class']!=0.0] = 1.0

# dropping NaNs from sex column (unchanged from fillna function bc non-numeric columns did not produce medians)
data_frame = data_frame.dropna()

# infer float64 object to be the better type for class column
data_frame = data_frame.infer_objects()
```
